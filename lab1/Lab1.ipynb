```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
import scipy.special as sc
#import seaborn as sns

uniform = rand.uniform(0,1, 1000)
normal01 = rand.normal(0,1,1000)
normal010 = rand.normal(0,10,1000)
beta22 = rand.beta(2,2,1000)
beta39 = rand.beta(3,9,1000)
x1, y1 = rand.multivariate_normal([0,0],[[1,0],[0,1]],10000).T
x2, y2 = rand.multivariate_normal([0,0],[[2,1],[1,2]],10000).T
```

```python
n, bins, patches = plt.hist(uniform, density=True)
y = 1/(1-0)
plt.hlines(y, xmin=0, xmax=1, color='y')
plt.show()
```

```python
n, bins, patches = plt.hist(normal01, density=True)
y = ((1/(np.sqrt(2*np.pi)*1))*np.exp(-(bins-0)**2/(2*1**2)));

plt.plot(bins,y)
plt.show()
```

```python
n, bins, patches = plt.hist(normal010, density=True)
y = ((1/(np.sqrt(2*np.pi)*10))*np.exp(-(bins-0)**2/(2*10**2)));

plt.plot(bins,y)
plt.show()
```

```python
n, bins, patches = plt.hist(beta22, density=True)
y = (((bins**(2-1))*(1-bins)**(2-1))/sc.beta(2,2));

plt.plot(bins,y)
plt.show()
```

```python
n, bins, patches = plt.hist(beta39, density=True)
y = (((bins**(3-1))*(1-bins)**(9-1))/sc.beta(3,9));

plt.plot(bins,y)
plt.show()
```

```python
n = plt.hist2d(x1,y1, bins=100, density=True)
plt.show()
```

```python
n = plt.hist2d(x2,y2, bins=500, density=True)
plt.show()
```
